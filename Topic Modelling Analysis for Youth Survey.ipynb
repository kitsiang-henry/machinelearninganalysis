**Elaboration about SMU as an Intergenerational Hub**

```python
import numpy as np
import pandas as pd
import nltk

from nltk.tokenize import word_tokenize

import re
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

import gensim
from gensim.models import LdaModel
from gensim.corpora import Dictionary
```

```python
youth_survey = pd.read_csv('youth_opinions_3.csv')
```

```python
change_name = {'Please elaborate for your choice in the previous question.':'Elaboration about SMU as an Intergenerational Hub'}
new_youth_survey = youth_survey.rename(columns=change_name)
opinion_elaboration = new_youth_survey['Elaboration about SMU as an Intergenerational Hub']
```

```python
tokenized_data = [word_tokenize(str(line)) for line in opinion_elaboration]
print(tokenized_data)
```

```python
WNL = WordNetLemmatizer()
others = ["", '...', '..', 'nan', 'nil', '\'s', 'elderly', 'think', 'also', 'sure', 'might', 'may', '’', 'idk', 'tbh']
stop_list = stopwords.words('english') + list(string.punctuation) + others

print(stop_list)
```

```python
# Text processing function
opinion_tokens = []

def expand(feedback):
    opinions = re.sub(r'n\'t', 'not', feedback)
    opinions = re.sub(r'\'ll', 'will', opinions)
    opinions = re.sub(r'\'m', 'am', opinions)
    return opinions

def process_opinions(text_opinions):
    for feedback in text_opinions:
        new_feedback = str(feedback).lower()
        new_feedback = expand(new_feedback)
        feedback_tokens = word_tokenize(new_feedback)
        feedback_tokens = [token for token in feedback_tokens if token not in stop_list]
        feedback_tokens = [WNL.lemmatize(token) for token in feedback_tokens]
        feedback_tokens = [token for token in feedback_tokens if token not in stop_list]
        opinion_tokens.append(' '.join(feedback_tokens))
    return opinion_tokens
```

```python
processed_elaboration = process_opinions(opinion_elaboration)
#print(processed_elaboration)

processed_elaboration_tokens = [word_tokenize(feedback) for feedback in processed_elaboration]
processed_elaboration_tokens = [i for i in processed_elaboration_tokens if i != []]
print(processed_elaboration_tokens)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

feedback_corpus = processed_elaboration
cv = CountVectorizer(feedback_corpus)
count_vect = cv.fit_transform(processed_elaboration)
print(count_vect)
```

```python
vocab = cv.get_feature_names()

cv_matrix = count_vect.toarray()

vect_df = pd.DataFrame(cv_matrix, columns=vocab)

word_counts = vect_df.sum(axis=0)

sorted_word_counts = word_counts.sort_values(ascending=False)
print(sorted_word_counts[:35])
```

```python
import gensim
from gensim.models import LdaModel
from gensim.corpora import Dictionary

dictionary = gensim.corpora.Dictionary(processed_elaboration_tokens)

print('Sample word to number mappings:', list(dictionary.items())[:15])
print('Total Vocabulary Size:', len(dictionary))

dictionary.filter_extremes(no_below=5, no_above=0.5)

print('Total Vocabulary Size after filters:', len(dictionary))
```

```python
corpus_feedback_words = [dictionary.doc2bow(text) for text in processed_elaboration_tokens]

print(corpus_feedback_words)
```

```python
# Topics
num_topics = 5
ldamodel = gensim.models.ldamodel.LdaModel(corpus_feedback_words, num_topics=num_topics, id2word=dictionary, passes=8, random_state=0)
```

```python
for num, topic in ldamodel.show_topics(formatted=True, num_topics=num_topics, num_words=10):
    print(str(num) + ': ' + topic + '\n')
```

```python
from gensim.models import CoherenceModel

log_perplexity = ldamodel.log_perplexity(corpus_feedback_words)
perplexity = 2**(-log_perplexity)
print('Perplexity: ',perplexity)

coherence_model_lda = CoherenceModel(model=ldamodel, texts=processed_elaboration_tokens, dictionary=dictionary, coherence='c_v')
coherence_lda = coherence_model_lda.get_coherence()
print('\nCoherence Score: ',coherence_lda)
```

**Benefits for branding SMU as an Intergenerational Hub**

```python
benefits_name = {'What benefits do you think SMU will get for branding itself as an Intergenerational Hub for youth and the elderly?': 'Benefits as Intergenerational Hub'}
new_youth_survey = new_youth_survey.rename(columns=benefits_name)
benefits = new_youth_survey['Benefits as Intergenerational Hub']
print(benefits)
```

```python
others = ['nil', '\'s', 'yes', 'smu', 'hahahaha', 'think', 'sure']
benefits_stop_list = stopwords.words('english') + list(string.punctuation) + others

print(benefits_stop_list)
```

```python
benefits_tokens = []

def expand(feedback):
    benefits = re.sub(r'n\'t', 'not', feedback)
    return benefits

def process_opinions(text_opinions):
    for feedback in text_opinions:
        new_feedback = str(feedback).lower()
        feedback_tokens = word_tokenize(new_feedback)
        feedback_tokens = [token for token in feedback_tokens if token not in benefits_stop_list]
        feedback_tokens = [WNL.lemmatize(token) for token in feedback_tokens]
        feedback_tokens = [token for token in feedback_tokens if token not in benefits_stop_list]
        benefits_tokens.append(' '.join(feedback_tokens))
    return benefits_tokens
```

```python
processed_benefits = process_opinions(benefits)
processed_benefits
```

```python
processed_benefits_tokens = [word_tokenize(token) for token in processed_benefits]
processed_benefits_tokens = [element for element in processed_benefits_tokens if element != []]
print(processed_benefits_tokens)
```

**Will SMU as an Intergenerational Hub disrupts students' lives?**

```python
new_name = {'Please elaborate for your choice in the previous question..1':'Elaboration about choice of disruption to students\' lives'}
new_youth_survey = new_youth_survey.rename(columns=new_name)
opinions_students_lives = new_youth_survey['Elaboration about choice of disruption to students\' lives']
opinions_students_lives
```

```python
others = ['\'s', 'nil', 'think', 'would', 'may', 'might', 'much', 'le', 'donot', 'ok', 'okay', '’']
students_lives_stop_list = stopwords.words('english') + list(string.punctuation) + others

print(students_lives_stop_list)
```

```python
students_lives_token = []

def expand(feedback):
    opinions = re.sub(r'n\'t', 'not', feedback)
    opinions = re.sub(r'\'ll', 'will', opinions)
    return opinions

def process_lives(students_lives):
    for line in students_lives:
        elaboration = line.lower()
        elaboration = expand(elaboration)
        elaboration = word_tokenize(elaboration)
        elaboration = [token for token in elaboration if token not in students_lives_stop_list]
        elaboration = [WNL.lemmatize(token) for token in elaboration]
        elaboration = [token for token in elaboration if token not in students_lives_stop_list]
        students_lives_token.append(' '.join(elaboration))
    return students_lives_token
```

```python
processed_students_lives = process_lives(opinions_students_lives)
processed_students_lives = [i for i in processed_students_lives if i != '']
processed_students_lives
```

```python
processed_students_lives_tokens = [word_tokenize(token) for token in processed_students_lives]
processed_students_lives_tokens
```

```python
from sklearn.feature_extraction.text import CountVectorizer

students_lives_corpus = processed_students_lives
cv = CountVectorizer(students_lives_corpus)
count_vect_sl = cv.fit_transform(processed_students_lives)
print(count_vect_sl)
```

```python
vocab_sl = cv.get_feature_names()
cv_matrix_sl = count_vect_sl.toarray()
vect_df_sl = pd.DataFrame(cv_matrix_sl, columns=vocab_sl)
word_counts_sl = vect_df_sl.sum(axis=0)
sorted_word_counts_sl = word_counts_sl.sort_values(ascending=False)
print(sorted_word_counts_sl)
```

```python
dictionary_sl = gensim.corpora.Dictionary(processed_students_lives_tokens)

print('Sample word to number mappings:', list(dictionary_sl.items())[:15])
print('Total Vocabulary Size:', len(dictionary_sl))

dictionary_sl.filter_extremes(no_below=3, no_above=0.5)

print('Total Vocabulary Size after filters:', len(dictionary_sl))

corpus_students_lives_words = [dictionary_sl.doc2bow(text) for text in processed_students_lives_tokens]
```
